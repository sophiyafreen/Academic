```python
from prophet import Prophet
import warnings
warnings.simplefilter(action='ignore',category=FutureWarning)

end1 = dt.date(end.year -1,end.month,end.day)
end2 = dt.date(end.year, end.month -6,end.day)
start1= dt.date(end1.year -3,end.month,end.day)

df1= reader.get_data_yahoo(Kryptolist,start1,end1)['Adj Close']
df2= reader.get_data_yahoo(Kryptolist,start1,end2)['Adj Close']

df1
df2
```

```python
modelfb1 = Prophet()
df1=df1.reset_index()
df1[['ds','y']] = df1[['Date','BTC-GBP']]
modelfb1.fit(df1)
```

```python
modelfb2=Prophet()
df2=df2.reset_index()
df2[['ds','y']] = df2[['Date','BTC-GBP']]
modelfb2.fit(df2)
```

```python
future = modelfb1.make_future_dataframe(periods=365)
```

```python
future1 = modelfb2.make_future_dataframe(periods=180)
```

```python
forecast1=modelfb1.predict(future)
forecast2=modelfb2.predict(future1)
```

```python
modelfb1.plot(forecast1)
plt.title('Backtest of BTC 1 year Prediction')
plt.show()
```

```python
modelfb2.plot(forecast2)
plt.title('Backtest of BTC 6 Month Prediction')
plt.show()
```

```python
import yfinance as yf
import pandas_datareader as reader
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import mean_squared_error

end=dt.datetime.now()
start=dt.date(end.year-4,end.month,end.day)

df3 = reader.get_data_yahoo('BTC-GBP',start,end)['Adj Close']


df3=df3.reset_index()
df3[['ds','y']] = df3[['Date','Adj Close']]

df3
```

```python

X=df3
size = int(len(X) * 0.7)
train, test = X[0:size], X[size:len(X)]
modelfb3 = Prophet()
modelfb3.fit(train)
future3 = modelfb3.make_future_dataframe(periods=len(test), freq='D')
forecasting = modelfb3.predict(future3)
y_true = test['y'].values
y_pred = forecasting.yhat.values[-len(test):]
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print('Test RMSE: %.3f' % rmse)

```

```python
forecasting.head()
```

```python
modelfb3.plot(forecasting)
plt.title('Prophet Forecast for BTC Price')
plt.ylabel('Price')
plt.xlabel('Date')
plt.show()
```
